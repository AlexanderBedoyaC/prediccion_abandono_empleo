```python
from google.cloud import bigquery
from google.oauth2 import service_account


credentials_path = 'clave/clave_db.json'
credentials = service_account.Credentials.from_service_account_file(credentials_path)
project_id = 'aplicaciones-analitica'

client = bigquery.Client(credentials= credentials, project=project_id) # creacion del client para trabajar en python
```

```python
consulta = 'select * from `aplicaciones-analitica.recursos_humanos.data_train`;'

data = client.query(consulta).to_dataframe()
```
